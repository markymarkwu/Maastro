```python
dose_patient1 = '/Users/wangyangwu/Documents/Maastro/NeuralNets/sample/P0439C0006I1473766/RTDOSE/1.3.6.1.4.1.32722.228.6.334797647287977966709364212462384620209'
dose_patient2 = '/Users/wangyangwu/Documents/Maastro/NeuralNets/sample/P0439C0006I2866736/RTDOSE/1.3.6.1.4.1.32722.228.6.193151678283566152558336891472184601865'
dose_patient3 = '/Users/wangyangwu/Documents/Maastro/NeuralNets/sample/P0439C0006I3038573/RTDOSE/1.3.6.1.4.1.32722.228.6.92322990415385122070377632268932077952'
dose_patients = [dose_patient1, dose_patient2, dose_patient3]
```

```python
import SimpleITK as sk
import numpy as np
CTs = []
DOSEs = []
for ct, dose in zip(CT_patients, dose_patients):
    file_ids = sk.ImageSeriesReader.GetGDCMSeriesIDs(str(ct))
    file_names = sk.ImageSeriesReader.GetGDCMSeriesFileNames(str(ct), file_ids[0])
    series_reader = sk.ImageSeriesReader()
    series_reader.SetFileNames(file_names)
    CT = series_reader.Execute()
    CTs.append(CT)
    
    file_ids = sk.ImageSeriesReader.GetGDCMSeriesIDs(str(dose))
    file_names = sk.ImageSeriesReader.GetGDCMSeriesFileNames(str(dose), file_ids[0])
    series_reader = sk.ImageSeriesReader()
    series_reader.SetFileNames(file_names)
    Dose = series_reader.Execute()
    DOSEs.append(Dose)  
```

### 2. Align CT and Dose by cropping the CT based on Dose origin so that both CT and dose has the same origin:

Take patient 2 for an example:

```python
CT_p2 = CTs[1]
Dose_p2 = DOSEs[1]
```

Origin difference is:

```python
origin_difference_p2 = np.array(CT_p2.GetOrigin())-np.array(Dose_p2.GetOrigin()[:3])
print(f"Origin difference: {origin_difference_p2}")
```

CT spacing is:

```python
spacing_p2 = np.array(CT_p2.GetSpacing())
print(f"CT spacing for patient 2 is: {spacing_p2}")
```

Get the correspoding slicing coordinates using "difference/spacing", here the result is floating but in order for indexing, I convert it into integer which gives rise to information loss:

```python
slicing_coor = (np.absolute(origin_difference_p2)/spacing_p2).astype(int)
print(f"slicing coordinates is: {slicing_coor}")
```

When converting to numpy array, (x,y,z) will be reversed, therefore pixel array will will have size (z,y,x), therefore we need to reverse slicing_coor:

```python
slicing_coor = slicing_coor[::-1]
print(f"Reversed slicing coordinates is: {slicing_coor}")
```

```python
CT_array = sk.GetArrayFromImage(CT_p2)
Dose_array = sk.GetArrayFromImage(Dose_p2)[0]
print(f"CT array size: {CT_array.shape}")
print(f"Dose array size: {Dose_array.shape}")
```

Slicing x from 16, slice y from 85, slice z from 43:

```python
Aligned_CT_array = CT_array[16:, 85:, 43:]
```
